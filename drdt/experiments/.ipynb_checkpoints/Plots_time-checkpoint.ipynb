```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
problems = ['AR', 'EAR', 'SR', 'ESR', 'AD', 'EAD']
```

# Soybean

```python
dataset_name = "Soybean"

n = 32
d = 5

NC = [9.092207193374634, 8.156677007675171, 338.32474875450134, 334.5960121154785, 226.04236698150635, 223.90353393554688]
NCgreedy = [15.298405885696411, 13.474026918411255, 379.5502018928528, 385.60985231399536, 261.28034114837646, 257.09931206703186]
Gr = [4.6694252490997314, 3.689176082611084, 381.22906494140625, 388.7180640697479, 300.71778893470764, 279.18977308273315]
```

```python
problem_indices = np.arange(len(problems))

plt.figure(figsize=(15,10))

offset = 0.1
x_offsets = np.array([-1.5*offset, -0.5*offset, 0.5*offset])

plt.vlines(problem_indices + x_offsets[0], 0, [a/100 for a in Gr], color='g', linewidth=15, label='Greedy')
plt.vlines(problem_indices + x_offsets[1], 0, [a/100 for a in NCgreedy], color='orange', linewidth=15, label='Node Cover Greedy')
plt.vlines(problem_indices + x_offsets[2], 0, [a/100 for a in NC], color='b', linewidth=15, label='Node Cover')

plt.xticks(problem_indices, problems, rotation=45, ha='right')

plt.title(f'Average Time for one sample {dataset_name}', fontsize=14)
plt.xlabel('Problems', fontsize=14)
plt.ylabel('Time', fontsize=14)
plt.legend(loc='best', fontsize=14)
plt.grid(True)
plt.show()
```
